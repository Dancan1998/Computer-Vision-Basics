```python
img = cv2.imread('DATA/crossword.jpg',0)
```

```python
plt.imshow(img, cmap='gray');
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap="gray")
```

```python
show_pic(img)
```

```python
ret, thresh1 =  cv2.threshold(img, 150,255,cv2.THRESH_BINARY)
show_pic(thresh1)
```

```python
thresh2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,8)
show_pic(thresh2)
```

```python
blended_img = cv2.addWeighted(src1=thresh1, alpha=0.6, src2=thresh2, beta=0.4, gamma=0)
show_pic(blended_img)
```
